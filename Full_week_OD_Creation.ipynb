```python
import pandas as pd
import itertools

# Load the two .dta files into DataFrames
od_pairs1 = pd.read_stata('/content/drive/MyDrive/SP/Card_table_from_all.dta')
od_pairs2 = pd.read_stata('/content/drive/MyDrive/SP/BC_Card_weekly_OD_10_17주 average num_total_in2020_ver4.dta')
```

```python
od_pairs1.columns
```

```python
od_pairs2.columns
```

```python
od_pairs1 = od_pairs1.loc[:, ['orig_sido', 'dest_sido', 'orig', 'dest', 'OD']]
od_pairs2 = od_pairs2.loc[:, ['orig_sido', 'dest_sido', 'orig', 'dest', 'OD']]
```

```python

# # Select specific columns
# od_pairs1 = od_pairs1.loc[:, ['orig_sido', 'dest_sido', 'orig', 'dest', 'OD']]
# od_pairs2 = od_pairs2.loc[:, ['orig_sido', 'dest_sido', 'orig', 'dest', 'OD']]

# Combine the two DataFrames and remove duplicates
od_pairs = pd.concat([od_pairs1, od_pairs2], ignore_index=True).drop_duplicates()

# Create a Series for weeks
weeks = pd.Series(range(19, 36), name='week')

# Create all combinations of weeks and OD pairs
combinations = list(itertools.product(weeks, od_pairs.itertuples(index=False, name=None)))

# Convert the combinations into a DataFrame
combined_data = pd.DataFrame(combinations, columns=['week', 'OD_tuple'])

# Split the OD_tuple into separate columns
combined_data[['orig_sido', 'dest_sido', 'orig', 'dest', 'OD']] = pd.DataFrame(combined_data['OD_tuple'].tolist(), index=combined_data.index)

# Drop the OD_tuple column as it's no longer needed
combined_data.drop(columns=['OD_tuple'], inplace=True)
```

```python
# add week_OD
combined_data['week_OD'] = combined_data['week'].astype(str) + combined_data['OD']
combined_data.to_csv("/content/drive/MyDrive/SP/Full_week_OD.csv")
```
